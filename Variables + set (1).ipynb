# Q1:Declare different variables for different types of data’s

```python
a = "You There,"
b = 10
c = 20.30
d = 2j
e = ["a", "b", "List"] 
f = ("c", "d", "Tuple")
g = {"e", "f", "Set"}
h = {"Idol" : "Rohit"}
i = False

print(f"{a} : {type(a)}")
print(f"{b} : {type(b)}")
print(f"{c} : {type(c)}")
print(f"{d} : {type(d)}")
print(f"{e} : {type(e)}")
print(f"{f} : {type(f)}")
print(f"{g} : {type(g)}")
print(f"{h} : {type(h)}")
print(f"{i} : {type(i)}")
```

# Q1:Write a Python program to create a set.

```python
newset = {1,2,3,4,5,6}
print(newset)
```

# Q2:Write a Python program to iteration over sets.

```python
newset1 = {'x','y',1,'z',2}
for i in newset1:
    print(i)
```

# Q3:Write a Python program to add a member(s) in a set.

```python
iset = {"apple","banana","cherry"}
yset = {"dates","elaichi"}
iset.update(yset)
print(iset)
```

# Q4:Write a Python program to remove item(s) from a given set.

```python
addthis = {'apple', 'banana', 'cherry', 'dates'}
delete = {'dates'}
addthis.difference_update(delete)
print(addthis)
```

# Q5:Write a Python program to remove an item from a set if it is present in the set

```python
myset4 = {'Hero', 'Honda', 'Yamaha', 'Bajaj'}
myset4.discard("Bajaj")
print(myset4)
```

# Q6:Write a Python program to find the maximum and the minimum value in a set.

```python
Mset = {3,5,62,23,12,1,4,6,7}
print("Maximum : ",max(Mset))
print("Minimum : ",min(Mset))
```

# Q7:Write a Python program to find the length of a set.
