# Q1: Write a Python script to add a key to a dictionary. 
Sample Dictionary : {0: 10, 1: 20}
Expected Result : {0: 10, 1: 20, 2: 30}

```python
dict1 = {
    0 : 10,
    1 : 20
}
print("Before update : ",dict1)
dict1[2] = 30
print("After update : ",dict1)
```

# Q2: Write a Python script to check whether a given key already exists in a dictionary

```python
dict2 = {0: 10, 1: 20, 2: 30 , 3: 40, 4: 50, 5: 60}
i = int(input("Enter the Key number you want to serach : "))
print(f"{i} key is present") if i in dict2 else print("{i} is not present")
```
