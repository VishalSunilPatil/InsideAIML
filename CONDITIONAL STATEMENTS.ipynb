# Check even or odd

#  Program to guess a number between 1 to 9 using elif 

```python
import random
b = random.randint(1, 10)
for i in range(9):
    a=int(input("Enter your guess : ")) 
    if a<1 or a>9:
        print("Enter number between range of 1 to 9.")
    elif a != b:
        print("Wrong guess.")
    else:
        print("Well guessed")
        break
```

# Write a Python program to count the number of even and odd numbers from a series of numbers 

```python
list = [1, 2, 3, 4, 5, 6]
Even = 0
Odd = 0
for num in list:
    if num%2 == 0:
        Even = Even + 1
    else:
        Odd = Odd + 1
print(f"Even numbers are {Even} and odd numbers are {Odd}")
```
