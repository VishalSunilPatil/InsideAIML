Q6: Create a function which can accept any number of parameters

```python
def Addition(*args):
    j = 0
    for i in args:
        j = j + i
    print(f"Addition is {j}.")
Addition(1)
Addition(1,6)    
Addition(1,6,10) 
```

Q9: Create a function which can find the largest item inside the list

```python
lst = [1,6,10,45]
def Highest(lst):
    max = lst[0]
    for i in lst:
        if i > max:
            max = i
    return max

print("Largest element in list:", Highest(lst))
```

Q10: Create a function which a print all even and odd number over a specified range

```python
s = int(input("Enter Starting Number : "))
e = int(input("Enter Last Number : "))
for i in range(s,e):
    if i%2==0:
        print(f"{i} is Even")
    else:
        print(f"{i} is Odd")
```
