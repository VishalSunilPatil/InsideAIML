# Q1: How tuples are different from the list

For List we use [] square braces to declare it,but while declaring Tuple () parenthesis is used.

List is mutable while Tuple is immutable.

List consume more memory while Tuple consumes less.

# Q5:Write a Python program to add an item in a tuple

Tuples are immutable

```python
newtuple3 = ('apple','banana')
print(newtuple3)
list1 = list(newtuple3)
list1.insert(1,'cherry')
newtuple3 = tuple(list1)
print(newtuple3)
```
