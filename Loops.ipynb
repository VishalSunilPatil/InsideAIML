# Create one list and iterate over the list using for,len and range 

```python
lst = [0, 1, 2, 3, 4, 5]
for e in range(len(lst)):
    print(e)
```

# Create a list fruits store apple,banana,grapes,oranges using for loop iterate and when it reaches to a banana if shoud print banana is available

```python
fruits = ["Apple","Banana","Cherry","Dates"]
for i in fruits:
    if i == "Banana":
        print("Banana is available")
    else:
        print(i)
```

# Convert the below english statement into code : values should be accepted from user If you got more than 60% marks overall I will give you salute five times(means we should print salute five times)else I will punish you five times(means we should print punishment five times) 

```python
Overall = float(input("Enter your Overall marks :"))
if Overall > 60:
    for i in range(5):
        print("Salute")
else:
    for i in range(5):
        print("Punishment")
```

# Convert the below english statement into code : values should be accepted from user If you got more than 80% marks in maths, and more than or equal to 70% marks in science and more than or equal to 90% marks in English then I will salute you as many times you want(means program should print whatever time user will say)

```python
Maths = int(input("Enter Maths mark: "))
Science = int(input("Enter Science mark: "))
English = int(input("Enter English mark: "))
if Science>=70 and Maths>80 and English>=90:
    s = int(input("I will salute you as may times you want. Enter the number of salutes : "))
    for i in range(s):
        print("Salute")
else:
    print("No Salute")
```

# Using for loop and range function print 0 – 100 but break the loop if it reaches 12 

```python
for g in range(101):
    if g == 12:
        break
    else:
        print(g)
```

# Using for loop and range function print 0 – 10 but it shouldn’t print 7 

```python
for h in range(0,11):
    if h == 7:
        continue
    else:
        print(h)
```

# Create a list of names and iterate over them and print the index and item present at that particular index

```python
Names = ["Rohit","Sachin","Virat","Mahi"]
for i in range(len(Names)):
    print(f"Index {i}",Names[i])
```

# From a range of numbers iterate over that range print which one is even and which one is odd 

```python
Lst = list(range(10))
for i in Lst:
    if i%2 == 0:
        print(f"{i} even")
    else:
        print(f"{i} odd")
```
